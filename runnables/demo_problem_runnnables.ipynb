```python
import random
class naklillm:
  def __init__(self):
    print('LLM created')

  def predict(self,prompt):
    response_list=[
        'delhi is the capital of india ',
        'london is the capital of england',
        'paris is the capital of france'
    ]
    return random.choice(response_list)


```

```python
llm=naklillm()
```

```python
class NakliPropmptemplate:
  def __init__(self,template,input_varibales):
    self.template=template
    self.input_varibales=input_varibales

  def format(self,input_dict):
    return self.template.format(**input_dict)
```

```python

template=NakliPropmptemplate(
    template='what is the capital of {country}',
    input_varibales=['country']
)
```

```python
prompt=template.format({'country':'france'})
```

```python

class NakliLLMchain:
  def __init__(self,llm,prompt):
    self.llm=llm
    self.prompt=prompt
  def run(self,input_dict):
    final_prompt= self.prompt.format(input_dict)
    result=self.llm.predict(final_prompt)
    return result
```

```python
template=NakliPropmptemplate(
    template='what is the capital of {country}',
    input_varibales=['country']
)
```

```python
llm=naklillm()
```

```python
chain=NakliLLMchain(llm,template)
```

```python
chain.run({'country':'france'})
```
